# Part 1

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#To show graph on ipynb
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11) #11 points from 0 to 5
y = x**2
```

```python
x
```

```python
y
```

```python
# THE FUNCTIONAL METHOD 
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#plt.show()
```

```python
# To create multiple plots on a same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method (a better way)
fig = plt.figure()

axes = fig.add_axes([0.1,.1,.8,.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([.2,.5,.4,.3])
#axes3 = fig.add_axes([.5,.2,.3,.4])

axes1.plot(x,y)
axes2.plot(y,x)

axes2.set_title('Tiny Plot')
axes1.set_title('Big Plot')
```

# Part 2

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
#plt.tight_layout() takes care of the overlapping plots

axes[0].plot(x,y)
axes[1].plot(y,x)

axes[0].set_title('Plot 0')
axes[1].set_title('Plot 1')

plt.tight_layout()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,3))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('Myfig.png',dpi = 200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label= 'X Squared')
ax.plot(x,x**3, label = 'X Cubed')

# ADDING LEGEND : 2 steps
ax.legend(loc = 0) # 0 means 'best location'
# ax.legend(loc = 1): upper right corner
# or use tuple loc = (0.1, 0.1)
```

# Part 3: Plot Appearance

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

#ax.plot(x,y,color='green')
ax.plot(x,y,color='green', linewidth=3,alpha=0.5,linestyle='-',marker='o', markersize=10, markerfacecolor='blue',
       markeredgewidth=5, markeredgecolor ='yellow') #RGB Hex Code for more custom color

# linewidth or lw
# alpha argument: for transparency
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',lw=2,ls='--')

# Set axes limit
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
